```python
import pandas as pd
import numpy as np
```

```python
unemployment_original = pd.read_csv('unemployment.csv')
natonality_immi_original = pd.read_csv('immigrants_by_nationality.csv')
age_immi_original = pd.read_csv('immigrants_emigrants_by_age.csv')
#ds3 = pd.read_csv ("immigrants_emigrants_by_age.csv")
#population_2017 = pd.read_csv ("pop2017.csv")
#population_2016 = pd.read_csv ("pop2016.csv")
#population_2015 = pd.read_csv ("pop2015.csv")
old_pop = pd.read_csv ("population.csv")
#list(population_original["Year"].unique())
#population_original = pd.concat([population_2017, population_2016, population_2015])
#population_original.columns=['Year', "District Code", "District Name" ,"Neighborhood Code" ,"Neighborhood Name", "Gender", "Number"]

age_immi_original.tail()
```

```python


#unemployment_original.astype({'Demand_occupation': 'str'}).dtypes
ages = list(old_pop["Age"].unique())
exclusion_age = ages [3:14]
print(ages[3:14])

```

```python

#clean_pop = population_original[(population_original['Neighborhood Code'] != 99)] 
clean_pop_old = old_pop[old_pop['Age'].isin(exclusion_age) & (old_pop['Neighborhood.Code'] != 99)] 


clean_unemployment = unemployment_original[(unemployment_original['Neighborhood Code'] != 99) &
                                           (unemployment_original['Month'] == "January") &
                                           (unemployment_original['Demand_occupation'] == "Registered unemployed")]


clean_immigrants_nat = natonality_immi_original[(natonality_immi_original['Neighborhood Code'] != 99) &
                                                (natonality_immi_original['Nationality'] != "No information") &
                                                (natonality_immi_original['Nationality'] != "Spain") &
                                               (natonality_immi_original['Number'] != 0)]
clean_immigrants_age = age_immi_original[age_immi_original['Age'].isin(exclusion_age) &
                                         (age_immi_original['Neighborhood Code'] != 99)]

clean_immigrants_nat.head()
```

```python
year = 2017

unemployment_year = clean_unemployment[clean_unemployment['Year']== year]

immigrants_year_age = clean_immigrants_age[clean_immigrants_age['Year']== year]       
immigrants_year_nat = clean_immigrants_nat[clean_immigrants_nat['Year']== year]               

#population_year = clean_pop[clean_pop['Year']== year]
population_year_old = clean_pop_old[clean_pop_old['Year']== year]

# Select duplicate rows except first occurrence based on all columns
#population_year_old  = population_year_old[population_year_old.duplicated()]


#population_year_old =  population_year_old[["Year","Neighborhood.Code","Neighborhood.Name","Gender","Age","Number"]].drop_duplicates()

immigrants_year_nat.head()
```

```python

immigrants_agg_nat = immigrants_year_nat.groupby(['Neighborhood Code',"Neighborhood Name","Nationality"], as_index=False).agg({'Number': "sum"})
immigrants_agg_nat = immigrants_agg_nat[immigrants_agg_nat["Number"] > 200]
immigrants_agg = immigrants_year_age.groupby(['Neighborhood Code',"Neighborhood Name"], as_index=False).agg({'Immigrants': "sum"})



#population_agg = population_year.groupby(['Neighborhood Code',"Neighborhood Name"], as_index=False).agg({'Number': "sum"})
pop_agg_old = population_year_old.groupby(['Neighborhood.Code',"Neighborhood.Name"], as_index=False).agg({'Number': "sum"})

unemployment_gender_agg = unemployment_year.groupby(['Neighborhood Code',"Neighborhood Name", "Gender"], as_index=False).agg({'Number': "sum"})

unemployment_agg = unemployment_year.groupby(['Neighborhood Code',"Neighborhood Name"], as_index=False).agg({'Number': "sum"})


immigrants_agg = immigrants_agg.rename(columns = {"Immigrants": "Total Immigrants"})

unemployment_agg = unemployment_agg.rename(columns = {"Number": "Registered Unemployed"})

population_agg = pop_agg_old.rename(columns = {"Number": "Total Residents", 'Neighborhood.Code': "Neighborhood Code","Neighborhood.Name" : "Neighborhood Name"})
#population_agg = population_agg.rename(columns = {"Number": "Total Residents"})


#immigrants_2017.tail()
#immigrants_2017_byhood.tail()
#hood1_pop = population_year [population_year["Neighborhood.Code"] ==1]
immigrants_agg_nat.head()
#unemployment_year = unemployment_original.drop(indexdrop , axis = 0 )
#unmployment_registere
#unemployment_agg.head()
```

```python
merged_unemp_immigr = unemployment_agg.merge(immigrants_agg, on= ['Neighborhood Code',"Neighborhood Name"])

database1 = merged_unemp_immigr.merge(population_agg, on = ['Neighborhood Code',"Neighborhood Name"])

database1["% Immigrants 2017"] = database1['Total Immigrants'] / database1["Total Residents"] * 100 

database1["% Unemployed 2017"] = database1['Registered Unemployed'] / database1["Total Residents"] * 100

database1.head()
```
