```python
# Importing the pandas and numpy libraries

import pandas as pd
import numpy as np
```

```python
# Reading the CSV files 

unemployment_original = pd.read_csv('unemployment.csv')

natonality_immi_original = pd.read_csv('immigrants_by_nationality.csv')

age_immi_original = pd.read_csv('immigrants_emigrants_by_age.csv')

old_pop = pd.read_csv ("population.csv")
```

```python
# creating the working age ranges

ages = list(old_pop["Age"].unique())
exclusion_age = ages [3:14]
print(ages[3:14])

```

```python
#cleaning the datasets using different coditions to get the information we need

clean_pop_old = old_pop[old_pop['Age'].isin(exclusion_age) & (old_pop['Neighborhood.Code'] != 99)] 


clean_unemployment = unemployment_original[(unemployment_original['Neighborhood Code'] != 99) &
                                           (unemployment_original['Month'] == "January") &
                                           (unemployment_original['Demand_occupation'] == "Registered unemployed")]


clean_immigrants_nat = natonality_immi_original[(natonality_immi_original['Neighborhood Code'] != 99) &
                                                (natonality_immi_original['Nationality'] != "No information") &
                                                (natonality_immi_original['Nationality'] != "Spain") &
                                               (natonality_immi_original['Number'] != 0)]


clean_immigrants_age = age_immi_original[age_immi_original['Age'].isin(exclusion_age) &
                                         (age_immi_original['Neighborhood Code'] != 99)]
```

```python
#filtering the data by year

year = 2017

unemployment_year = clean_unemployment[clean_unemployment['Year']== year]

immigrants_year_age = clean_immigrants_age[clean_immigrants_age['Year']== year]       
immigrants_year_nat = clean_immigrants_nat[clean_immigrants_nat['Year']== year]               


population_year_old = clean_pop_old[clean_pop_old['Year']== year]

```

```python
#gruping the data sets by Neighborhood Code, adding and renaming the informantion we want

immigrants_agg_nat2017 = immigrants_year_nat.groupby(['Neighborhood Code',"Neighborhood Name","Nationality"]
                                                     , as_index=False).agg({'Number': "sum"})

immigrants_agg = immigrants_year_age.groupby(['Neighborhood Code',"Neighborhood Name"]
                                             , as_index=False).agg({'Immigrants': "sum"})

pop_agg_old = population_year_old.groupby(['Neighborhood.Code',"Neighborhood.Name"]
                                          , as_index=False).agg({'Number': "sum"})

unemployment_gender_agg2017 = unemployment_year.groupby(['Neighborhood Code',"Neighborhood Name", "Gender"]
                                                        , as_index=False).agg({'Number': "sum"})

unemployment_agg = unemployment_year.groupby(['Neighborhood Code',"Neighborhood Name"]
                                             , as_index=False).agg({'Number': "sum"})

immigrants_agg = immigrants_agg.rename(columns = {"Immigrants": "Total Immigrants"})

unemployment_agg = unemployment_agg.rename(columns = {"Number": "Registered Unemployed"})

population_agg = pop_agg_old.rename(columns = {"Number": "Total Residents",
                                               'Neighborhood.Code': "Neighborhood Code",
                                               "Neighborhood.Name" : "Neighborhood Name"})

```

```python
#merging the data sets by Neighborhood Code, and later to merge wit the other years

merged_unemp_immigr = unemployment_agg.merge(immigrants_agg, on= ['Neighborhood Code',"Neighborhood Name"])

database1 = merged_unemp_immigr.merge(population_agg, on = ['Neighborhood Code',"Neighborhood Name"])

database1["% Immigrants 2017"] = database1['Total Immigrants'] / database1["Total Residents"] * 100 

database1["% Unemployed 2017"] = database1['Registered Unemployed'] / database1["Total Residents"] * 100

database1.head()
```

```python
# and now we repeat teh process for year 2016

year = 2016

unemployment_year = clean_unemployment[clean_unemployment['Year']== year]

immigrants_year_age = clean_immigrants_age[clean_immigrants_age['Year']== year]       
immigrants_year_nat = clean_immigrants_nat[clean_immigrants_nat['Year']== year]               

population_year_old = clean_pop_old[clean_pop_old['Year']== year]

immigrants_year_nat.head()
```

```python

immigrants_agg_nat_2016 = immigrants_year_nat.groupby(['Neighborhood Code',"Neighborhood Name","Nationality"]
                                                      , as_index=False).agg({'Number': "sum"})

immigrants_agg = immigrants_year_age.groupby(['Neighborhood Code',"Neighborhood Name"]
                                             , as_index=False).agg({'Immigrants': "sum"})

unemployment_gender_agg_2016 = unemployment_year.groupby(['Neighborhood Code',"Neighborhood Name"
                                                          , "Gender"], as_index=False).agg({'Number': "sum"})

unemployment_agg = unemployment_year.groupby(['Neighborhood Code',"Neighborhood Name"]
                                             , as_index=False).agg({'Number': "sum"})

immigrants_agg = immigrants_agg.rename(columns = {"Immigrants": "Total Immigrants 2016"})

unemployment_agg = unemployment_agg.rename(columns = {"Number": "Registered Unemployed 2016"})

population_agg = pop_agg_old.rename(columns = {"Number": "Total Residents 2016",
                                               'Neighborhood.Code': "Neighborhood Code",
                                               "Neighborhood.Name" : "Neighborhood Name"})

population_agg.head()
```

```python
#merging the data sets by Neighborhood Code, and later to merge with the previous years

merged_unemp_immigr = unemployment_agg.merge(immigrants_agg, on= ['Neighborhood Code',"Neighborhood Name"])

database2 = merged_unemp_immigr.merge(population_agg, on = ['Neighborhood Code',"Neighborhood Name"])

database2["% Immigrants 2016"] = database2['Total Immigrants 2016'] / database2["Total Residents 2016"] * 100 

database2["% Unemployed 2016"] = database2['Registered Unemployed 2016'] / database2["Total Residents 2016"] * 100

database = database1.merge(database2, on = ['Neighborhood Code',"Neighborhood Name"])

database2.head()
```

```python
# and now we repeat the process for year 2015

year = 2015

unemployment_year = clean_unemployment[clean_unemployment['Year']== year]

immigrants_year_age = clean_immigrants_age[clean_immigrants_age['Year']== year]       
immigrants_year_nat = clean_immigrants_nat[clean_immigrants_nat['Year']== year]               

population_year_old = clean_pop_old[clean_pop_old['Year']== year]

immigrants_year_nat.head()
```

```python

immigrants_agg_nat2015 = immigrants_year_nat.groupby(['Neighborhood Code',"Neighborhood Name","Nationality"]
                                                     , as_index=False).agg({'Number': "sum"})

immigrants_agg = immigrants_year_age.groupby(['Neighborhood Code',"Neighborhood Name"]
                                             , as_index=False).agg({'Immigrants': "sum"})

pop_agg_old = population_year_old.groupby(['Neighborhood.Code',"Neighborhood.Name"]
                                          , as_index=False).agg({'Number': "sum"})

unemployment_gender_agg_2015 = unemployment_year.groupby(['Neighborhood Code',"Neighborhood Name", "Gender"]
                                                         , as_index=False).agg({'Number': "sum"})

unemployment_agg = unemployment_year.groupby(['Neighborhood Code',"Neighborhood Name"]
                                             , as_index=False).agg({'Number': "sum"})


immigrants_agg = immigrants_agg.rename(columns = {"Immigrants": "Total Immigrants 2015"})

unemployment_agg = unemployment_agg.rename(columns = {"Number": "Registered Unemployed 2015"})

population_agg = pop_agg_old.rename(columns = {"Number": "Total Residents 2015"
                                               , 'Neighborhood.Code': "Neighborhood Code"
                                               ,"Neighborhood.Name" : "Neighborhood Name"})
```

```python
#merging the data sets by Neighborhood Code, and later to merge with the previous years

merged_unemp_immigr = unemployment_agg.merge(immigrants_agg, on= ['Neighborhood Code',"Neighborhood Name"])

database3 = merged_unemp_immigr.merge(population_agg, on = ['Neighborhood Code',"Neighborhood Name"])

database3["% Immigrants 2015"] = database3['Total Immigrants 2015'] / database3["Total Residents 2015"] * 100 

database3["% Unemployed 2015"] = database3['Registered Unemployed 2015'] / database3["Total Residents 2015"] * 100

database = database.merge(database3, on = ['Neighborhood Code',"Neighborhood Name"])

database3.describe()
```
