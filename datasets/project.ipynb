```python
# Importing the pandas library as pd

import pandas as pd
```

```python
# Reading the CSV files 

unemployment = pd.read_csv('unemployment.csv')
unemployment.head (20)
```

```python
immigrants_nat = pd.read_csv('immigrants_by_nationality.csv')
immigrants_nat.head(20)
```

```python
# Do column titles make sense?

print("Titles of unemployment dataset: ",list(unemployment))
print("--------------"*8)
print("Titles of immigrants by nationality dataset: ",list(immigrants_nat))
```

```python
# Merge datasets

combined = pd.merge(unemployment, immigrants_nat, on='Neighborhood Code', how='inner')
```

```python
combined.head()
```

```python
# Drop columns that are not interesting for the analysis and save in a new dataframe

combined_new = combined.drop(['Month', 'District Code_x','District Name_x','Neighborhood Name_x'], axis=1)
combined_new.head()
```

```python
# Rename column titles

combined_new = combined_new.rename(columns={'Year_x': 'Year_Unemployment', 'Year_y': 'Year_Immigrant','District Code_y':'District Code', 'Neighborhood Name_y':'Neighborhood Name','Number_y':'Number_Immigrants','Number_x':'Number_Unemployed'})
```
